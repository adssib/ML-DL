### Ndarray objects

Numpy provides the
[ndarray](https://docs.scipy.org/doc/numpy/reference/generated/numpy.ndarray.html)
type, which stands for "N-dimensional array." The simplest is
one-dimensional:

```python
y = [2.0, 4.0, 6.0]        # Create a list object with three slots, each refering to an integer object
x = np.array(y)      # Create an ndarray object, initializing each x[i] by copying the value from y[i]
print("the x array is", x)

w = np.array([[2, 4, 6],   # Create a 2x3 int32 ndarray, initialized from a nested list
              [5, 3, 1]])
print(f"the w array is {w}")
```

We can check the type of this new object:

```python
print(type(x))             # Variable 'x' refers to an object of type 'ndarray' defined by the numpy module
```

An ndarray object has attributes that tell us about its dimensions and its data type:

```python
print(x.ndim)       # The 'ndim' (number of dimensions) on this particular ndarray object is 1
print(len(x))       # The length of the first dimension (the only dimension) is 3
print(x.shape)      # The 'shape' tuple contains the length of each dimension (all 'ndim' of them)
print(x.dtype)      # The 'dtype' (data type) on this particular ndarray object is 32-bit integer
```

You can create a two-dimensional (2D) ndarray object from nested lists:



```python
x = np.array([[2, 4, 6], [5, 3, 1]])   # Create a nested list (3 list objects, 6 integers) and

print(x)                                     # use it to initialize a new ndarray object
print(x.ndim)          # The 'ndim' (number of dimensions) of this ndarray object is 2
print(x.shape)
```

###  Indexing and Slicing

Numpy [indexing and
slicing](https://www.numpy.org/devdocs/user/quickstart.html#indexing-slicing-and-iterating)
both work similarly to lists:

```python
x = np.array([2, 4, 6, 8, 10])  # Create an int32 ndarray, initialized from three integer objects
print(x[0])                     # Index [0] has the int32 value 2
print(x[-1])                    # Index [len(x)-1] has the int32 value 10
print(x[2:])                    # Slice [1:len(x)] creates a new ndarray array skipping first item
print(x[:2])
print(x[2:4])
print(x[:-1])
```

### Key Functions:

Matplotlib is used for creating static, animated, and interactive visualizations in Python. Below are some key functions used to create and enhance visualizations:

- **plt.[figure()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html)**: Creates a new figure for plotting. This is useful for generating multiple plots in a single session or setting custom figure properties such as size and resolution.

- **plt.[plot()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)**: Creates line plots, ideal for visualizing continuous data and trends over time or another variable.

- **plt.[scatter()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html)**: Generates scatter plots to represent relationships between two variables by displaying individual data points.

- **plt.[hist()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html)**: Plots histograms to visualize the distribution of a dataset, helping to understand the frequency of data ranges.

- **plt.[grid()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html)**: Adds a grid to the plot, making it easier to interpret data points relative to the axes.

- **plt.[legend()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html)**: Displays a legend on the plot, which is useful for labeling different data series or elements in the visualization.

- **plt.[xlabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xlabel.html)**: Labels the x-axis of the plot, providing context for the data displayed along the horizontal axis.

- **plt.[ylabel()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.ylabel.html)**: Labels the y-axis of the plot, providing context for the data displayed along the vertical axis.

- **plt.[title()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.title.html)**: Adds a title to the plot to give a concise description of what the visualization represents.

- **plt.[annotate()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html)**: Adds annotations, such as text or arrows, to specific points on the plot to highlight important data features.

- **plt.[show()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.show.html)**: Displays the plot window. This is necessary to render and visualize the plot you have created.



```python
# Plot sine and cosine over the interval [-2Ï€, 2Ï€]
x = np.linspace(-2 * np.pi, 2 * np.pi, num=100)
sin_x = np.sin(x)
cos_x = np.cos(x)

plt.figure(figsize = (5,3))
plt.plot(x, sin_x, label = "Sin(x)")
plt.plot(x, cos_x, label = "Cos(x)")

# X and Y limits
#plt.xlim(0,np.pi)
#plt.ylim(-2.0,2.0)

# Grid and legend display
plt.grid()
plt.legend()

# Set x and y labels
plt.xlabel("x")
plt.ylabel("F(x)")

# Annotate specific data point
plt.annotate(
  "my point", xy=(np.pi, -1), xytext=(1, -0.75),
  arrowprops={"facecolor": "magenta"}
)
plt.title("Trig Functions"); # use ; to supress printing the output text
```

```python
# Set figure size
plt.figure(figsize=(6, 4))

# Set marker, marker size and color
plt.plot(
  x, sin_x, label="Sin", linestyle = ":",
  marker="o", markersize=5, color="navy", markerfacecolor = "yellow"
)

# Set line style and width
plt.plot(
  x, cos_x, label="Cos",
  color="orange", linestyle="--", linewidth=5
)

# Set x-axis limits
plt.xlim(-np.pi,np.pi)

# Set y-axis limits
plt.ylim(-1, 1)

# Enable grid lines
plt.grid()

# Hide top and right spines/axes
plt.gca().spines["top"].set_visible(False)
plt.gca().spines["right"].set_visible(False)

# Set legend location and number of columns
plt.legend(loc="upper left", ncol=2)
plt.xlabel("x")
plt.ylabel("y")

# Change font size
plt.title("Trig functions", fontsize=18);
```

```python
x = np.arange(0,20)
y = x**2
u = np.sqrt(x)

plt.figure(figsize=(3,3))
plt.plot(x, y,
         label = "Quadratic Function",
         linestyle = "-.",
         color = "green",
         linewidth = 3
         )
plt.plot(x, u,
         label = "Square Root Function",
         linestyle=":",
         color = "blue",
         linewidth = 2)

plt.ylim(0,10)
plt.xlim(0,10)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Math Functions")
plt.legend(loc = "upper right")
plt.grid()
```

### Multiple subplots

```python
# Create new figure
plt.figure(figsize=(8, 4))

# First subplot
plt.subplot(2, 2, 1)  # Rows, columns, index
plt.plot(x, sin_x, label="$sin(x)$")
plt.legend()
plt.title("Sine")

# Second subplot
plt.subplot(2, 2, 2)
plt.plot(x, cos_x, label="$cos(x)$")
plt.legend()
plt.title("Cosine")

# Third subplot spanning 2 columns
plt.subplot(2, 1, 2)
plt.plot(x, cos_x + sin_x, label="$cos^2(x)$")
plt.legend()
plt.title("Cosine + Sine")

# Fix vertical margin to avoid label overlap
plt.subplots_adjust(hspace=0.5)
```

# 3. PyTorch

In PyTorch, a tensor is essentially an n-dimensional array. You can create a tensor from a Python list or a NumPy array using `torch.tensor()`.

### Difference Between Arrays (NumPy) and Tensors (PyTorch)

- **Arrays** from **NumPy**, are used for general scientific computing. On the other hand, **Tensors** from **PyTorch** are designed for deep learning.

- **NumPy Arrays** support CPU-only computations. However, **PyTorch Tensors** can be run on **GPU** for faster computations.


Let's import torch!

```python
import torch
```

### Creating tensors

```python
# Create a tensor from a Python list
t = torch.tensor([1, 2, 3, 4])
print(t)
print(type(t))
```

You can create tensors with random values drawn from a uniform or normal distribution.

```python
# Create a tensor with random values (uniform distribution)
random_tensor = torch.rand(3, 3)  # 3x3 tensor with values in range [0, 1)
print(random_tensor)

# Create a tensor with random values from normal distribution
normal_tensor = torch.randn(3, 3)  # Standard normal distribution (mean=0, std=1)
print(normal_tensor)
```

### Important Note: PyTorch and NumPy Functions

Many of the functions you are familiar with in **NumPy** (such as `np.arange()`, `np.linspace()`, and others) have direct counterparts in **PyTorch**. In fact, PyTorch is designed to be very similar to NumPy, which makes transitioning between them straightforward.

However, there are some **key differences** to keep in mind:

- **PyTorch functions** (like `torch.tensor()`, `torch.zeros()`, etc.) should be applied **only to tensors**.
- **NumPy functions** (like `np.array()`, `np.zeros()`, etc.) should be applied **only to NumPy arrays**.

Mixing them up can cause errors or unexpected behavior in your code. Always ensure that you are applying the correct library's functions to the appropriate data types.

Pay close attention to the data type you are working with:
- If you're working with **tensors**, use **PyTorch functions**.
- If you're working with **NumPy arrays**, use **NumPy functions**.
